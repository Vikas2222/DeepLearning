A Simple logistic regression in theano


```python
import numpy
import theano
import theano.tensor as T
```

```python
rng = numpy.random
print rng

N = 400
feats = 784
```

```python
D = (rng.randn(N,feats), rng.randint(size=N, low=0, high=2))
training_steps = 100
print D # 400*784
```

```python
x = T.dmatrix("xdata")
y = T.dvector("y")
print 
```

```python
w = theano.shared(rng.randn(feats), name="w")
print rng.randn(feats).shape
```

```python
b = theano.shared(0., name='b')
print 
```

```python
print ("Initial model :")
print (w.get_value())
print (b.get_value())
# bias and weights
```

```python
p_1 = 1/(1 + T.exp(-T.dot(x,w) - b))
prediction = p_1 > 0.5
xent = -y * T.log(p_1) - (1-y) * T.log(1-p_1)
cost = xent.mean() + 0.01 * (w ** 2).sum()
gw, gb = T.grad(cost, [w, b])

print p_1
print prediction
print xent
print cost
print gw
print gb
```

```python
train = theano.function(
                        inputs = [x, y],
                        outputs = [prediction, xent],
                        updates = ((w, w - 0.1 *gw), (b, b -0.1 * gb)))

predict = theano.function(inputs=[x], outputs=prediction)
                                                    
```

```python
for i in range(training_steps):
    pred, err = train(D[0], D[1])
    
print ("Final Model : ")
print w.get_value()
print(b.get_value())
print("target values for D:")
print(D[1])
print("prediction on D:")
print(predict(D[0]))
```
